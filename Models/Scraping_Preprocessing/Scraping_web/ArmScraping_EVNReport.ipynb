```python
import numpy as np
import pandas as pd

import requests # to request the context of page
from bs4 import BeautifulSoup #to convert to html parser readable

from pprint import pprint
import time
import re
```

```python
list_links = []
for i in range(2,8):
    url = f"https://evnreport.com/arm/series-category/raw-unfiltered-arm/page/{i}/"
    response=requests.get(url)
    page=response.content
    page=BeautifulSoup(page,'html.parser')
    hashtags_html=page.find_all('article')
    for n in np.arange(0,len(hashtags_html)):
        link = hashtags_html[n].find('a')['href']
        list_links.append(link)
```

```python
hashtags = []
for  i in range(0, len(list_links)):
    url = list_links[i]
    response=requests.get(url)
    page=response.content
    page=BeautifulSoup(page,'html.parser')
    hashtags_html=page.find_all('p')
    hashtags.append([i.get_text() for i in hashtags_html]) # get text using list comphrahensions
```

```python
hashtags = [item for sublist in hashtags for item in sublist]
```

```python
hashtags = [re.sub('\xa0', '', i) for i in hashtags]
hashtags = [re.sub('\xad', '', i) for i in hashtags]
hashtags = [re.sub('\r', '', i) for i in hashtags]
hashtags = [re.sub('\n', '', i) for i in hashtags]
hashtags = [re.sub('\t', '', i) for i in hashtags]
hashtags = [re.sub('-', '', i) for i in hashtags]
hashtags = re.sub("[\(\[\«].*?[\)\]\»]", " ", hashtags)
hashtags = re.sub("[^\u0561-\u0587\u0531-\u0556\ ։]+",' ',hashtags)
```

```python
hashtags = [re.sub(' ×', '', i) for i in hashtags]
hashtags = [re.sub('Δ', '', i) for i in hashtags]
```

```python
hashtags = [re.sub('————————', '', i) for i in hashtags]
```

```python
redundant_words = [ 'ԱՋԱԿՑՈՒԹՅՈՒՆԱՆԿԱԽԼՐԱԳՐՈՒԹՅՈՒՆ', 'Your Message*', 'Ձեր էլփոստի հասցեն չի հրապարակվելու։ Պարտադիր դաշտերը նշված են *ով',
 'Անուն * ',
 'Էլփոստի հասցե * ',' @ 2021 EVN Report. Բոլոր իրավունքները պաշտպանված են ',
                  ' Save my name, email, and website in this browser for the next time I comment. ',
                  'Բաժանորդագրվել ', ''
                  ]
```

```python
hashtags = [y for y in hashtags if y not in redundant_words]
```

```python
hashtags = "".join(hashtags)
```
